```python
from datetime import datetime
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def convString2Date(t):
    return datetime.strptime(t, '%d/%m/%Y %H:%M')
def convString2Float(f):
    if f == 'NA':
        return float(0)
    else :
        return float(f.replace(',','.'))
```

```python
data_path = '/Users/jeremielequeux/Documents/Git/deep_weather/RW_results_data/Germany-2014-2016.csv'
```

```python
data = np.loadtxt(open(data_path, "rb"),
                  dtype={'names': ('Time', 'Power'),'formats': ('datetime64[m]','f8')},
                  delimiter=";",
                  skiprows=1,
                  converters = {0:  convString2Date, 1: convString2Float})
```

# Plot one particular day

```python
print data['Time'][20064]
print data['Time'][20160]
```

```python
slice_day = np.empty([96], dtype={'names': ('Time', 'Power'),'formats': ('datetime64[m]','f8')})
slice_day['Time'] = data['Time'][20064:20160]
slice_day['Power'] = data['Power'][20064:20160]
```

```python
plot = plt.plot(slice_day['Time'],slice_day['Power'])
```
